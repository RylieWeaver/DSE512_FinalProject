## I) On the fly dataset

Randomly samples fixed-length chunks from a FASTA file on-the-fly.

```python
VOCAB = {'A': 0, 'T': 1, 'G': 2, 'C': 3, 'N': 4}
DATA_PATH = r'download.20260314.151350\Phytozome\PhytozomeV12\early_release\Ptrichocarpa_444_v3.1\assembly\Ptrichocarpa_444_v3.0.fa.gz'

data  = pyfastx.Fasta(DATA_PATH)
names   = [s.name for s in data]
lengths = [len(s)  for s in data]   
total   = sum(lengths)
weights = [l / total for l in lengths]

def get_random_chunk(chunk_size=8000):
    name   = random.choices(names, weights=weights, k=1)[0]
    seq    = data[name]
    start  = random.randint(0, len(seq) - chunk_size)
    tokens = [VOCAB.get(b.upper(), 4) for b in seq[start:start+chunk_size].seq]
    return torch.tensor(tokens, dtype=torch.long)

chunk = get_random_chunk()
print(chunk.shape)
```
